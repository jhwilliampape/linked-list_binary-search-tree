## Stacks and Queues (Review)

** Stacks ** as the name suggests is a data structure that allows for data to follow the Last In First Out priciple(LIFO). Think of a stack of pancakes for example. To get the first pancake you would  start with the top and go down.

##### Searching through a stack will be Linear Time O(n) - Constant Space O(1)
##### Selecting the last item will be done in Constant Time O(1) - Constant Space O(1)
##### Adding to the stack should take Constant Time O(1) - Constant Space O(1)

** Queues ** are similar but in this case follow the First In First Out principle(FIFO). Think of this as a line in a black friday sale. The first person camped out for the big screen tv is the first to get it.

##### Searching through a queue will be Linear Time O(n) - Constant Space O(1)
##### Selecting the first item will be done in Constant Time O(1) - Constant Space O(1)
##### Adding to the queue should take Constant Time O(1) - Constant Space O(1)

```python
# Python lists function as stacks (if we want them to)

stack = []

stack.append(10) # O(1) to add to the list
stack.append(20)
stack.append(30)

#removal of the final item(pop())
# removal from the end of list.
# We can also pop at a specific index - if we do that, this becomes a linear operation
last_item = stack.pop() #built-in pop function
print(last_item)

#searching through looping

for i in stack:
    print(i)
    
#know about the existence of queues
#Built-in queue functionality (collection deques)

first_item = stack.pop()

first_item
```

## Linked List (Data Structure)

A linked list is created by using the node class. We create a Node object and create another class to use this node object. We pass the appropriate values thorugh the node object to point the to the next data elements.

There are some advantages and disadvantages with this data structure. **Advantages** Linked Lists can save memory because they can be flexibile with memory management which saves memory. **Disadvantages** Finding or adding to the list requires traversing the entire list.

```python
class LinkedListNode:
    def __init__(self, value):
        self.value = value
        self.next = None
        
    def traverseList(self):
        node = self
        while node != None:
            print(node.value)
            node = node.next

node1 = LinkedListNode('Mon')
node2 = LinkedListNode('Tue')
node3 = LinkedListNode('Wed')

node1.next = node2
node2.next = node3

print(node1.traverseList())

```

```python
# A complete implementation of a linked list
# 2 components to this solution -- a Node class and a linkedList class

class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def pushOn(self,new_value):
        new_node = Node(new_value)
        new_node.next = self.head
        self.head = new_node
        
    def insertAfter(self, prev_node, new_value):
        #Check if the previous node even exists
        if prev_node is None:
            print('The given previous node must not be empty')
            return
        #If the node is not empty, then create a new Node instance
        new_node = Node(new_value)
        
        #Update the previous nodes old pointer to point to my new Node value
        new_node.next = prev_node.next
        
        prev_node.next = new_node
        
    #This method will append to the END of the linked list
    def append(self, new_value):
        #Create a new Node with new value
        new_node = Node(new_value)
        
        #Check if linked list is empty
        # and if it is, make this new node the head node(beginning of the list)
        if self.head is None:
            self.head = new_node
            
        #BUT if the list is NOT empty - traverse through the end
        # and add a new value to end of list
        last = self.head
        #while last.next is NOT None - continue to loop until we
        # find a None value for next pointer
        while last.next:
            last = last.next
            
        # Change our current last node value to point at our new node
        last.next = new_node
        
    def traverse(self):
        temp = self.head
        #While temp is NOT None = keep looping through the links in
        # this list until you reach a None value
        while temp:
            print(temp.value)
            temp = temp.next
            
weekdays_links = LinkedList()

#Insert a new day into LL

weekdays_links.pushOn('Mon')
weekdays_links.append('Tue')
weekdays_links.insertAfter(weekdays_links.head.next,'Wed')
weekdays_links.traverse()


        
```

## Binary Search Trees

```python
# Why? very quick look-up and insertion -
# child nodes on the left must be less than parent node
# child nodes on right must be greater than parent node

class BST:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
    def insert(self, value):
        if value < self.value:
            if self.left is None:
                self.left = BST(value)
            else:
                self.left.insert(value)
        else:
            if self.right is None:
                self.right = BST(value)
            else:
                self.right.insert(value)
        return self
    
    def contains(self, value):
        """
        This method accepts an integer for the value parameter
        The value parameter is what we are looking for inside of our BST
        """
        if value < self.value: #if passed in value is less than current value
            if self.left is None:
                return False
            else:
                return self.left.contains(value)
        elif value > self.value:
            if self.right is None:
                return False
            else:
                return self.right.contains(value)
        else:
            return True
    
    #Getting Min or Max values
    def getMinValue(self):
        if self.left is None:
            return self.value
        else:
            return self.left.getMinValue()
    def getMaxValue(self):
        if self.right is None:
            return self.value
        else:
            return self.right.getMaxValue()
        
        
        
    def remove(self, value, parent=None):
        if value < self.value:
            if self.left is not None:
                self.left.remove(value, self)
        elif value > self.value:
            if self.right is not None:
                self.right.remove(value, self)
                
        else:
            if self.left is not None and self.right is not None:
                self.value = self.right.getMinValue()
                self.right.remove(value, self)
            elif parent is None:
                if self.left is not None:
                    self.value = self.left.value
                    self.left = self.left.left
                    self.right = self.left.left
                elif self.right is not None:
                    self.value = self.right.value
                    self.left = self.right.left
                    self.right = self.right.right
                else:
                    self.value = None
            elif parent.left == self:
                parent.left = self.left if self.left is not None else self.right
            elif parent.right == self:
                parent.right = self.left if self.left is not None else self.right
        return self
        

bst_example = BST(39)
bst_example.insert(40)
bst_example.insert(50)
print(bst_example.getMaxValue())
print(bst_example.contains(40))
bst_example.remove(40)
print(bst_example.contains(40))
        
```

#### Problem 2: Binary Search Tree

Using the above examples as a guide, create your own interpretation of the a Binary Search Tree class. You can not use the code above exactly, but again it can be used as a guide. This problem requires you to think about how a Binary Search Tree works and create one using your own logic.

*Remember* Binary Search Trees start with a head node and each node to the left of that will be smaller, each node to the right of it will be greater. The far left node should be the lowest number(if one exists) that is available. The far right node (if one exists) should be the greatest number

```python
class BST:
    def __init__(self, var):
        self.var = var
        self.los = None
        self.his = None
    
    def insert(self, var):
        if var < self.var:
            if self.los is None:
                self.his = BST(var)
            else:
                self.los.insert(var)
        else:
            if self.his is None:
                self.his = BST(var)
            else:
                self.his.insert(var)
        return self

    def contains(self, var):
        """
        This method accepts an integer for the value parameter
        The value parameter is what we are looking for inside of our BST
        """
        if var < self.var:        #if passed in value is less than current value
            if self.los is None:
                return False
            else:
                return self.los.contains(var)
        elif var > self.var:
            if self.his is None:
                return False
            else:
                return self.his.contains(var)
        else:
            return True
        
    #Getting Min or Max values
    def getLowest(self):
        if self.los is None:
            return self.var
        else:
            return self.los.getLowest()
    def getHighest(self):
        if self.his is None:
            return self.var
        else:
            return self.his.getHighest()
        
        
        
    def remove(self, var, parent=None):
        if var < self.var:
            if self.los is not None:
                self.los.remove(var, self)
        elif var > self.var:
            if self.his is not None:
                self.his.remove(var, self)
                
        else:
            if self.los is not None and self.his is not None:
                self.var = self.his.getLowest()
                self.his.remove(var, self)
            elif parent is None:
                if self.los is not None:
                    self.var = self.los.var
                    self.los = self.los.los
                    self.his = self.los.los
                elif self.his is not None:
                    self.var = self.his.var
                    self.los = self.his.los
                    self.his = self.his.his
                else:
                    self.var = None
            elif parent.los == self:
                parent.los = self.los if self.los is not None else self.his
            elif parent.his == self:
                parent.his = self.los if self.los is not None else self.his
        return self
        

bst_example = BST(39)
bst_example.insert(40)
bst_example.insert(50)
print(bst_example.getHighest())
print(bst_example.contains(40))
bst_example.remove(40)
print(bst_example.contains(40))
```

```python
#Find Smallest Number  - (Without min)
#Create a function that given a list of numbers (non-sorted) find the lowest number in the list and return it.
#Example Input: [50,30,4,2,11,0]
#Example Output: 0
#Example Input 2: [40,3,4,2]
#Example Output 2: 2

def lowest_number(numbers):
    lowest = None
    for num in numbers:   # Loop through each number in the list
        if not lowest:
            lowest = num 
        if num < lowest: # Update lowest if current number is less
            lowest = num
    return lowest

lowest_number([50,30,4,2,11,0])
```
